```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("data/hits_data.csv",delimiter=";", 
                 #index_col="row_num"
                )
```

```python
df.rename({'session_durantion':'session_duration'}, axis=1, inplace=True)
print(df.shape)
df.head()
```

Spliting the dataset into test and train, based on inavailablity of hits data

```python
train = df[df['hits']!="\\N"]
test = df[df['hits']=="\\N"]
```

```python
train[['hits']]  = train[['hits']].astype(int)
print(train.shape)
train.head()
```

```python
print(test.shape)
test.head()
```

```python
train.isnull().sum()
```

It is seen that the path_id_set columns have missing data. 

Find the Unique Ids and check if they repeat in train and test

```python
path_id_train= train.loc[train["path_id_set"].notnull(),"path_id_set"].str.split(';').tolist()
ids = {}
ids_dummy = {}
i=10000
for path_split in path_id_train:
    for split in path_split:
        if split in ids:
            ids[split]+=1
        else:
            ids[split]=1
            ids_dummy[split] =i
            i+=1
print("Unique Ids with train data:",len(ids))
```

```python
path_id_test= test.loc[test["path_id_set"].notnull(),"path_id_set"].str.split(';').tolist()
for path_split in path_id_test:
    for split in path_split:
        if split in ids:
            ids[split]+=1
        else:
            ids[split]=1
            ids_dummy[split] =i
            i+=1
print("Unique Ids with train and test data:",len(ids))
```

This shows that the path_id_set values might keep increasing and is not a constant set of categories. I had further analysed and found that only less than 1% of the dataset had more than 5 path_id_sets(and so be considered as outliers). So below I am converting the path_id_set into 6 different columns, with first 5 of oringinal values splitted. The 6th column will have the remaining number of path_ids as I did not want to waste the data.

```python
path = train["path_id_set"].str.split(';', n=5,expand=True)
path['no_of_ids']=train['path_id_set'].str.split(';')
path['no_of_ids']=path['no_of_ids'].str.len()
path.drop(axis=1, columns=5,inplace=True)
path.columns = ['path_1','path_2','path_3','path_4','path_5','no_of_ids'] 
path.head()
```

The path values are mapped with values ranging from 10000-35000 so that the values will scale better. The scaling is done in the 'Trivago_ML_feature_engg.ipynb' file.

```python
path['path_1'] = path['path_1'].map(ids_dummy)
path['path_2'] = path['path_2'].map(ids_dummy)
path['path_3'] = path['path_3'].map(ids_dummy)
path['path_4'] = path['path_4'].map(ids_dummy)
path['path_5'] = path['path_5'].map(ids_dummy)
path.head()
```

```python
train[['path_1','path_2','path_3','path_4','path_5','no_of_ids']] = path[['path_1','path_2','path_3','path_4','path_5','no_of_ids']]
train.drop(axis=1, columns='path_id_set',inplace=True)
```

```python
cols=['row_num', 'locale', 'day_of_week', 'hour_of_day', 'agent_id',
       'entry_page', 'traffic_type', 'session_duration',
        'path_1', 'path_2', 'path_3', 'path_4', 'path_5', 'no_of_ids','hits']
train = train[cols]
train.head()
```

```python
#train.no_of_ids.fillna(0,inplace=True)
train.fillna(0,inplace=True)
train.head()
```

```python
train[['path_1','path_2','path_3','path_4','path_5']]=train[['path_1','path_2','path_3','path_4','path_5']].astype(int)
```

```python
train.dtypes
```

From above its seen that session duration is saved as an object inspite of being numbers. After analysing it I found there was some missing data as '\N'. 
This was seen due to the crossing over of "hour_of_day" to the next day(the deduction was made by analysing the max "session_duration"<(24-"hour_of_day")x60x60). Thus whenever a persons usage started in one day and crossed over to next day in same session it became NaN

```python
# hour_0_sess_N = (train["hour_of_day"]==0) & (train["session_duration"]=='\\N')
# train.loc[hour_0_sess_N,"session_duration"] = train.loc[train['session_duration']!='\\N','session_duration'].astype('int').mean()
hour_n0_sess_N = (train["session_duration"]=='\\N')
train.loc[hour_n0_sess_N,"session_duration"] = (24-train.loc[hour_n0_sess_N, "hour_of_day"])*3600
train[['session_duration']]  = train[['session_duration']].astype(int)
```

So, what I have done is converted the remaining hours of the day to seconds and saved as the unknown 'session_duration' value. It can be even more, I cannot be sure. But this should work fine.

```python
days_of_week = ['Monday', 'Tuesday','Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
locales = ['L1','L2', 'L3','L4', 'L5', 'L6']
```

```python
day_of_week_dict = {}
locale_dict = {}
for i,a in enumerate(locales):
    locale_dict[a] = i+1
for i,a in enumerate(days_of_week):
    day_of_week_dict[a] =i+1
print(day_of_week_dict)
print(locale_dict)
```

```python
train['day_of_week'] = train['day_of_week'].map(day_of_week_dict)
train['locale'] = train['locale'].map(locale_dict)
train.head()
```

```python
agent = {10:1, 9:2, 1:3, 8:4, 2:5, 13:6, 7:7, 14:8}
train['agent_id'] = train['agent_id'].map(agent)
train.fillna(-1,inplace=True)
train[['agent_id']]=train[['agent_id']].astype(int)
```

```python
entry={2113: 1, 2116:2, 2100:3, 2111:4, 2114:5, 2700:6, 8001:7, 2115:8, 2226:9, 2316:10}
train['entry_page'] = train['entry_page'].map(entry)
train.fillna(-1,inplace=True)
train[['entry_page']]=train[['entry_page']].astype(int)
```

The day of week, locale, entry and agent values have been mapped with dummy values for now just for my convenience. It will be properly handled in the 'Trivago_ML_feature_engg.ipynb' file.
